```python
# Importar Pandas e Matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importação dos dados
df = pd.read_csv('data/raw.csv')
```

```python
# Exibir os tipos de dados de cada coluna para verificar a necessidade de atuação
df.dtypes
```

```python
# Transformando a coluna 'Data ultimo aumento' para datetime
df['Data ultimo aumento'] = pd.to_datetime(df['Data ultimo aumento'])
```

```python
# Separando colunas numéricas e categóricas
numerical_cols = [col for col in df.columns if df[col].dtype != 'object' and df[col].dtype != 'datetime64[ns]']
categorical_cols = [col for col in df.columns if df[col].dtype == 'object']
```

```python
# Resumo das colunas numéricas
df[numerical_cols].describe()
```

```python
# Resumo das colunas categóricas
df[categorical_cols].describe()
```

```python
# Verificar se existem valores faltantes nas colunas
df.isna().any()
```

```python
# Verificar a quantidade de valores faltantes em cada coluna
df.isna().sum()
```

```python
# Tratar os valores faltantes em Area Formação por: "Não informado":
df['Area Formacao'].fillna('Não informado', inplace=True)
```

```python
# Verificar se existem valores duplicados
df.duplicated().sum()
```

```python
# Contagem de valores por categoria de 'Nivel'
df['Nivel'].value_counts()
```

```python
# Contagem de valores por categoria de 'Faixa Salarial'
df['Faixa Salarial'].value_counts()
```
